# Subtraction from data collected 04/05/2025

```python
#import data with boron shield
data_boron = pd.read_csv(("data/boron_04052023.txt"),dtype=np.float64)
boron_bins = data_boron.iloc[:,0]
boron_cts = data_boron.iloc[:,1]
boron_bins = np.asarray(boron_bins,dtype=np.double)
boron_cts = np.asarray(boron_cts)

#import data taken without boron
data_bknd = pd.read_csv("data/bknd_03312023.txt")
bknd_bins = data_bknd.iloc[:,0]
bknd_cts = data_bknd.iloc[:,1]
bknd_bins = np.asarray(bknd_bins,dtype=np.double)
bknd_cts = np.asarray(bknd_cts)
```

```python
#subtract background
from math import*
#subtraction
sbtr_boron=((boron_cts)-(bknd_cts))
print (sbtr_boron)
#error margin
sigma_cts = np.sqrt(((bknd_cts)**2)+((boron_cts)**2))
error = sigma_cts
print (sigma_cts)
```

```python
#create 1d plot for bknd
fig,axes = plt.subplots(1,1,figsize=(9.0,8.0),sharex=True)
ax1 = axes

xmin=1
xmax=10000
stepsize=100

ax1.step(bknd_bins,bknd_cts,color='m',where='mid', linestyle='-',label='background', linewidth=1)



ymin = 1.0
ymax = 2000.0

ax1.set_yscale('linear')
ax1.set_xlim(xmin, xmax) 
ax1.set_ylim(ymin,ymax)
ax1.set_xlabel(r'keV')
ax1.set_ylabel('counts')
ax1.legend(loc=1,prop={'size':22})


for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.tight_layout()
plt.show()
```

```python
import lmfit as lf
from scipy.interpolate import InterpolatedUnivariateSpline as interp

def gaussian(x,a,m,s):
    return a*np.exp(-(x-m)**2/(2*s**2))

gauss_mod = lf.Model(gaussian)
gauss_params = gauss_mod.make_params()
gauss_params.add('s',value=15,min=1)

fine = np.arange(min(bknd_bins),max(bknd_bins))
```

```python
interpolated = interp(bknd_bins,bknd_cts,k=5)
interpolatey = interpolated(fine)
```

```python
a=3900
arange=15
gauss_params.add('m',value=a,vary=True)
gauss_params.add('a',value=interpolatey[fine==a][0],vary=True)
fita_bk = gauss_mod.fit(interpolatey[abs(fine - a) < arange],gauss_params,x=fine[abs(fine - a) < arange])
print(fita_bk.best_values)

b=6250
brange=20
gauss_params.add('m',value=b,vary=True)
gauss_params.add('a',value=interpolatey[fine==b][0],vary=True)
fitb_bk = gauss_mod.fit(interpolatey[abs(fine - b) < brange],gauss_params,x=fine[abs(fine - b) < brange])
print(fitb_bk.best_values)
```

```python
#set up a 1d plot
fig,axes = plt.subplots(1,1,figsize=(9.0,8.0),sharex=True)
ax1 = axes

blue = '#118DFA'
xmin=0
xmax=10000
stepsize=100
#ax1.xaxis.set_ticks(np.arange(xmin, xmax, stepsize))
#ax1.step(bknd_bins,bknd_cnts, where='mid',color='m', linestyle='-', \
#            label='background', linewidth=2)
ax1.step(bknd_bins,bknd_cts,color='m',where='mid', linestyle='-',label='background', linewidth=1)
plt.plot(fine[abs(fine - a) < arange],fita_bk.best_fit,color='green',linewidth=3)
plt.plot(fine[abs(fine - b) < brange],fitb_bk.best_fit,color='green',linewidth=3)


ymin = 1.0
ymax = 3000.0


ax1.set_yscale('linear')
#ax1.set_yscale('log')
ax1.set_xlim(xmin, xmax) 
ax1.set_ylim(ymin,ymax)
ax1.set_xlabel(r'bins')
ax1.set_ylabel('counts')
ax1.legend(loc=1,prop={'size':22})


for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.tight_layout()
plt.show()
```
