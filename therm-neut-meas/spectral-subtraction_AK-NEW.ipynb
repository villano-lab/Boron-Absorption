```python
#import libraries
import numpy as np
import pandas as pd

#import matplotlib for plotting
import matplotlib as mpl
from matplotlib import pyplot as plt
plt.style.use('../mplstyles/stylelib/standard.mplstyle')
```

```python
#import data with boron shield
data_boron = pd.read_csv("data/borondata_09152022.txt")
boron_bins = data_boron.iloc[:,0]
boron_cts = data_boron.iloc[:,1]
boron_bins = np.asarray(boron_bins,dtype=np.double)
boron_cts = np.asarray(boron_cts)

#import data taken without boron
data_bknd = pd.read_csv("data/noboron_data09122022.txt")
bknd_bins = data_bknd.iloc[:,0]
bknd_cts = data_bknd.iloc[:,1]
bknd_bins = np.asarray(bknd_bins,dtype=np.double)
bknd_cts = np.asarray(bknd_cts)
```

```python
#subtract background
from math import*
#subtraction
sbtr_boron=((boron_cts)-(bknd_cts))
print (sbtr_boron)
#error margin
sigma_cts = np.sqrt(((bknd_cts)**2)+((boron_cts)**2))
print (sigma_cts)
```

```python
#create 1d plot for bknd
fig,axes = plt.subplots(1,1,figsize=(9.0,8.0),sharex=True)
ax1 = axes

xmin=100
xmax=16000
stepsize=100

ax1.step(bknd_bins,bknd_cts,color='m',where='mid', linestyle='-',label='background', linewidth=1)



ymin = 1.0
ymax = 3500.0

ax1.set_yscale('linear')
ax1.set_xlim(xmin, xmax) 
ax1.set_ylim(ymin,ymax)
ax1.set_xlabel(r'keV')
ax1.set_ylabel('counts')
ax1.legend(loc=1,prop={'size':22})


for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.tight_layout()
plt.show()
```
